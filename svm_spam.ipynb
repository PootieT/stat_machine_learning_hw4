```python
# parameter search for kernalized method
Cvals = [0.01,0.03,0.1,0.3,1,3,10,30]
# 
sigma_vals = [0.01,0.03,0.1,0.3,1,3,10,30]

count = 0

for sigma in sigma_vals:
    K = np.array([utils.hamming_kernel(x1,x2,sigma) for x1 in X_train for x2 in X_train]).reshape(X_train.shape[0],X_train.shape[0])
    print K.shape, K[:20,:10]
    scaler = preprocessing.StandardScaler().fit(K)
    scaleK = scaler.transform(K)
    print scaleK[:20, :10]
    KK = np.vstack([np.ones((K.shape[0],)),scaleK.T]).T
    for C in Cvals:
        svm = LinearSVM_twoclass()
        svm.theta = np.zeros((KK.shape[1],))
        svm.train(KK,yy_train,learning_rate=1e-2,reg=C,num_iters=13000,verbose=True,batch_size=KK.shape[0])
        Kval = np.array([utils.hamming_kernel(x1,x2,sigma) for x1 in X_val for x2 in X_train]).reshape(X_val.shape[0],X_train.shape[0])
        scaler = preprocessing.StandardScaler().fit(Kval)
        scaleKval = scaler.transform(Kval)
        scaleKKval = np.vstack([np.ones((Kval.shape[0],)),scaleKval.T]).T
        error = sum((svm.predict(scaleKKval)!=yy_val)*1.0)/float(len(yy_val))
        count += 1
        print "trial ",str(count),", sigma: ",str(sigma),", C: ",str(C),", error: ",str(error)
```
